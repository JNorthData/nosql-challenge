### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
results = list(establishments.find({'scores.Hygiene': 20}))

# Use count_documents to display the number of documents in the result
ct = establishments.count_documents({'scores.Hygiene': 20})
print(f"Number of Documents found: {ct}")
```

```python
# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(results)
```

```python
# Display the number of rows in the DataFrame
print(f"Number of rows in DataFrame: {len(hygiene_df)}")
```

```python
# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments in London with a Rating Value >= 4
hi_rating = list(establishments.find({
    '$and': [
        {'RatingValue': {'$gte': 4}},
        {'$or': [{'AddressLine2': 'London'}, {'AddressLine3': 'London'}]}
    ]
}))

# Use count_documents to display the number of documents in the result
ct = establishments.count_documents({
    '$and': [
        {'RatingValue': {'$gte': 4}},
        {'$or': [{'AddressLine2': 'London'}, {'AddressLine3': 'London'}]}
    ]
})
print(f"Number of Documents found: {ct}")
```

```python
# Display the first document in the results using pprint
pprint(hi_rating[0])
```

```python
# Convert the result to a Pandas DataFrame
hi_rating_df = pd.DataFrame(hi_rating)

# Display the number of rows in the DataFrame
print(f"Number of rows in DataFrame: {len(hi_rating_df)}")
```

```python
# Display the first 10 rows of the DataFrame
hi_rating_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the 
# latitude and longitude of 'Penang Flavours'.
query = list(establishments.find({
    # Rating value must equal 5
    "RatingValue": 5,
    "geocode.latitude": {
        "$gte": 51.490142 - 0.01,
        "$lte": 51.490142 + 0.01
    },
    "geocode.longitude": {
        "$gte": 0.08384 - 0.01,
        "$lte": 0.08384 + 0.01
    }
# Sort by hygiene score
}).sort("scores.Hygiene"))
```

```python
# Convert result to Pandas DataFrame
query_df = pd.DataFrame(query)

query_df.head(10)
```
